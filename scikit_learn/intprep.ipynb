```python
# supervised learning
from sklearn.linear_model import LinearRegression

x = [[1],[2],[3],[4]]
y = [100,200,300,400]

model = LinearRegression()
model.fit(x,y)

print(model.predict([[5]]))
```

```python
# unsupervised learning

from sklearn.cluster import KMeans

X = [[1,2],[1,4],[5,6],[6,8],[7,8],[0,0]]

model = KMeans(n_clusters=2)
model.fit(X)

print(model.labels_)
```

```python
from sklearn.linear_model import LogisticRegression

X = [[1], [2], [3], [4], [5]]
y = [0, 0, 1, 1, 2]

model = LogisticRegression()
model.fit(X, y)

print(model.predict([[5]]))
```

```python
# accuracy metrics

from sklearn.metrics import accuracy_score

y_true = [0,1,1,0]
y_pred = [0,1,1,1]

print(accuracy_score(y_true, y_pred))
```

```python

from sklearn.metrics import mean_squared_error
import numpy as np

y_true = [3,5,7]
y_pred = [2.5,5.5,6.5]

rmse = np.sqrt(mean_squared_error(y_true, y_pred))
print(rmse)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_true = [0, 1, 1, 0, 1]
y_pred = [0, 1, 0, 0, 1]

print("Accuracy :", accuracy_score(y_true, y_pred))
print("Precision:", precision_score(y_true, y_pred))
print("Recall   :", recall_score(y_true, y_pred))
print("F1 Score :", f1_score(y_true, y_pred))
```

```python

from sklearn.metrics import roc_auc_score

y_true = [0, 1, 1, 0, 1]
y_prob = [0.1, 0.9, 0.6, 0.3, 0.8]

print("ROC-AUC:", roc_auc_score(y_true, y_prob))
```

```python

import numpy as np

# inputs
x = np.array([1.2, 2.3, 3.1])

# weights
w = np.array([0.5, -0.2, 0.7])

# bias
b = 0.1

# activation (ReLU)
def relu(z):
    return max(0, z)

z = np.dot(w, x) + b
output = relu(z)

print("Neuron output:", output)
```

```python

import numpy as np

# forward pass
def predict(x, w):
    return x * w   # simple neuron

# loss (MSE)
def loss(y_true, y_pred):
    return (y_true - y_pred) ** 2

# backprop (adjust weight)
def update_weight(w, x, y_true, y_pred, lr=0.1):
    dL_dw = -2 * x * (y_true - y_pred)  # gradient
    return w - lr * dL_dw

x = 2
y_true = 10
w = 1  # initial weight

for i in range(5):
    y_pred = predict(x, w)
    L = loss(y_true, y_pred)
    w = update_weight(w, x, y_true, y_pred)
    print(f"Iteration {i}, Loss: {L}, Weight: {w}")
```
