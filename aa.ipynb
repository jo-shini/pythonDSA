```python
def domainGet(email):
    domain = email.split("@")
    return domain[1]


domainGet("user@domain.com")
```

```python
def findDog(word):
    if "dog" in word:
        return True
    return False


findDog("Is there a dog here?")
```

```python
def countDog(w):
    c = 0
    ws = w.split()
    for i in ws:
        if "dog" == i:
            c += 1
    return c


countDog("This dog runs faster than the other dog dude!")
```

```python
def caught_speeding(speed, is_birthday):
    offset = 5 if is_birthday == True else 0
    if speed <= 60 + offset:
        print("No Ticket")
    elif speed <= 80 + offset:
        print("Small Ticket")
    else:
        print("Big Ticket")


caught_speeding(81, True)
```
