```python
class Solution:
    def twoSum(self, nums: list[int], target: int) -> list[int]:
        # print(len(nums))
        for i in range(len(nums)):
            for j in range(1, len(nums)):
                if (nums[i] + nums[j]) == target and (i != j):
                    return [i, j]


s = Solution()
s.twoSum([3, 3], 6)
```

```python
class Solution:
    def removeDuplicates(self, nums: list[int]) -> int:
        if not nums:
            return 0

        i = 0
        for j in range(1, len(nums)):
            if nums[j] != nums[i]:
                i += 1
                nums[i] = nums[j]

        return i + 1


s = Solution()
s.removeDuplicates([0, 0, 1, 1, 1, 2, 2, 3, 3, 4])
```

```python
class Solution:
    def flipAndInvertImage(self, image: list[list[int]]) -> list[list[int]]:
        reverList = []
        for i in image:
            reverList.append(i[::-1])

        reverList = [[1 if x == 0 else 0 for x in sublist] for sublist in reverList]

        return reverList


s = Solution()
s.flipAndInvertImage([[1, 1, 0], [1, 0, 1], [0, 0, 0]])
```
