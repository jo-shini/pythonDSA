```python
class Solution:
    def romanToInt(self, s: str) -> int:
        dictx = {"I": 1, "V": 5, "X": 10, "L": 50, "C": 100, "D": 500, "M": 1000}

        splitting = list(s)
        res = 0

        i = 0
        while i < len(splitting) - 1:
            if dictx[s[i]] < dictx[s[i + 1]]:
                subx = dictx[s[i + 1]] - dictx[s[i]]
                res += subx
                i += 2
            else:
                res += dictx[s[i]]
                i += 1

        if i == len(s) - 1:
            res += dictx[s[i]]

        return res


s = Solution()
s.romanToInt("XV")
```

```python
class Solution:
    def maxFreqSum(self, s: str) -> int:
        vowels = set("aeiou")
        freq = {}

        for ch in s:
            freq[ch] = freq.get(ch, 0) + 1

        max_vowel = 0
        max_consonant = 0

        for ch, cnt in freq.items():
            if ch in vowels:
                if cnt > max_vowel:
                    max_vowel = cnt
            else:
                if cnt > max_consonant:
                    max_consonant = cnt

        return max_vowel + max_consonant


s = Solution()
s.maxFreqSum("successes")
```

```python
from typing import List
from collections import Counter


class SolutionDP:
    def countMaxOrSubsets(self, nums: List[int]) -> int:
        # 1) Compute the maximum OR value
        target = 0
        for x in nums:
            # print(target |= x)
            # print("before", "x = ", x, "target = ", target, "XOR", x | target)
            target |= x
            # print("after", "x = ", x, "target = ", target, "XOR", x | target)
        # print(target)

        # 2) dp[v] = how many subsets produce OR exactly v
        dp = Counter()

        # 3) Process each number and update OR counts
        for x in nums:
            new_dp = Counter()

            # Start a new subset with just x
            new_dp[x] += 1
            # print(new_dp, dp)

            # Extend all existing ORs with x
            for v, c in dp.items():
                # print(v, c)
                new_dp[v | x] += c

            # Merge into dp (adds counts for same keys)
            dp += new_dp
            # print(dp)

        # # 4) The answer is how many subsets yield OR == target
        return dp[target]


s = SolutionDP()
s.countMaxOrSubsets([3, 2, 1, 5])
```

```python
class Solution:
    def alternatingSum(self, nums: list[int]) -> int:
        res = 0
        for i in range(len(nums)):
            if i & 1 == 1:
                res -= nums[i]
            else:
                res += nums[i]
        return res


s = Solution()
s.alternatingSum([1, 3, 5, 7])
```
