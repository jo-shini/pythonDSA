```python
class Solution:
    def twoSum(self, nums: list[int], target: int) -> list[int]:
        # print(len(nums))
        for i in range(len(nums)):
            for j in range(1, len(nums)):
                if (nums[i] + nums[j]) == target and (i != j):
                    return [i, j]


s = Solution()
s.twoSum([3, 3], 6)
```

```python
class Solution:
    def removeDuplicates(self, nums: list[int]) -> int:
        if not nums:
            return 0

        i = 0
        for j in range(1, len(nums)):
            if nums[j] != nums[i]:
                i += 1
                nums[i] = nums[j]

        return i + 1


s = Solution()
s.removeDuplicates([0, 0, 1, 1, 1, 2, 2, 3, 3, 4])
```

```python
class Solution:
    def diStringMatch(self, s: str) -> list[int]:
        res = []
        low = 0
        high = len(s)
        for i in s:
            if i == "I":
                res.append(low)
                low += 1
            else:
                res.append(high)
                high -= 1
        res.append(low)
        return res


s = Solution()
s.diStringMatch("DDI")
```

```python
class Solution:
    def generate(self, numRows: int) -> list[list[int]]:
        res = []
        for i in range(numRows):
            row = [1]
            if i > 0:
                prev = res[-1]
                for k in range(1, i):
                    row.append(prev[k - 1] + prev[k])
                row.append(1)
            res.append(row)
        return res


s = Solution()
s.generate(3)
```

```python
class Solution:
    def romanToInt(self, s: str) -> int:
        dictx = {"I": 1, "V": 5, "X": 10, "L": 50, "C": 100, "D": 500, "M": 1000}

        splitting = list(s)
        res = 0

        i = 0
        while i < len(splitting) - 1:
            if dictx[s[i]] < dictx[s[i + 1]]:
                subx = dictx[s[i + 1]] - dictx[s[i]]
                res += subx
                i += 2
            else:
                res += dictx[s[i]]
                i += 1

        if i == len(s) - 1:
            res += dictx[s[i]]

        return res


s = Solution()
s.romanToInt("XV")
```

```python
class Solution:
    def longestCommonPrefix(self, strs: list[str]) -> str:
        if not strs:
            return ""

        shortest = min(strs, key=len)
        for i, ch in enumerate(shortest):
            # print(i, ch)
            for s in strs:
                # print(s)
                if s[i] != ch:
                    return shortest[:i]
        return shortest


s = Solution()
s.longestCommonPrefix(["flower", "flow", "flight"])
# s.longestCommonPrefix(["dog", "racecar", "car"])
```

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        # spaceleft = s.split(" ")
        # joinwithoutspace = "".join(spaceleft).lower()
        onlyaplha = ""
        for i in s:
            if i.isalnum():
                onlyaplha += i
        return onlyaplha.lower() == onlyaplha[::-1].lower()


s = Solution()
s.isPalindrome("A man, a plan, a canal: Panama")
```

```python
class Solution:
    def moveZeros(self, nums: list[int]) -> None:
        left = 0
        for i in range(len(nums)):
            if nums[i] != 0:
                nums[left], nums[i] = nums[i], nums[left]
                left += 1
        return nums


s = Solution()
s.moveZeros([0, 1, 0, 3, 12, 22, 0])
```

```python
class Solution:
    def palindrome(self, s: str) -> bool:
        i, j = 0, len(s) - 1
        while i < j:
            if s[i] != s[j]:
                return False
            i += 1
            j -= 1
        return True


s = Solution()
s.palindrome("dw")
```

```python
class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:
        merge = ""
        if len(word2) >= len(word1):
            for i in range(len(word2)):
                if i < len(word1):
                    merge += word1[i]
                merge += word2[i]

        else:
            for i in range(len(word1)):
                merge += word1[i]
                if i < len(word2):
                    merge += word2[i]
        return merge


s = Solution()
s.mergeAlternately("abcd", "pq")
```

```python
class Solution:
    def canMakeArithmeticProgession(self, arr: list[int]) -> bool:
        arr.sort()
        diff = arr[-1] - arr[-2]
        for i in range(len(arr) - 1):
            if arr[i + 1] - arr[i] != diff:
                return False
        return True


s = Solution()
s.canMakeArithmeticProgession([1, 2, 4])
```

```python
class Solution:
    def scoreOfString(self, s: str) -> int:
        listx = list(s)
        s = 0
        for i in range(len(listx) - 1):
            s += abs(ord(listx[i]) - ord(listx[i + 1]))
        return s


s = Solution()
s.scoreOfString("zaz")
```
