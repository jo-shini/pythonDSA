```python
class Solution:
    def maximumWealth(self, accounts:list[list[int]])->int:
        maxi = 0
        for account in accounts:
            totalwealth = sum(account)
            if totalwealth > maxi:
                maxi = totalwealth
        return maxi

s=Solution()
s.maximumWealth([[1,5],[7,3],[3,5]])
```

```python
class Solution:
    def countConsistentStrings(self, allowed:str, words:list[str])->int:
        total = 0
        allowedstring = set(allowed)
        for word in words:
            for ch in word:
                if ch not in allowedstring:
                    break
            else:
                total+=1
        return total

        
s=Solution()
s.countConsistentStrings("cad", ["cc","acd","b","ba","bac","bad","ac","d"])
```

```python
class Solution:
    def interpret(self, command:str)->str:
        # splitCommad = list(command)
        result = []
        comm = 0
        while comm < len(command):
            if command[comm] == "(":
                comm += 1
                if command[comm] == ")":
                    result.append("o")
                    comm += 1
                else:
                    result.append("al")
                    comm += 3
            else:
                result.append("G")
                comm += 1
        return "".join(result)

        return command.replace("()","o").replace("(al)","al")


s=Solution()
s.interpret("(al)G(al)()()G")
```

```python
class Solution:
    def convertDateToBinary(self, date:str)->str:
        binary_date = []
        split_date = date.split('-')
        for i in split_date:
            binary_date.append(bin(int(i))[2:])
        return "-".join(binary_date)
s=Solution()
s.convertDateToBinary("1900-01-01")
```

```python
class Solution:
    def numberofBeams(self, bank:list[str])->int:
        prev = 0
        total = 0
        
        for row in bank:
            count = row.count('1')
            if count > 0:
                total += prev * count
                prev = count

        return total

s=Solution()
s.numberofBeams(["011001","000000","010100","001000"])
```

```python
class Codec:
    def encode(self, longUrl:str)-> str:
        """Encodes a URL to a shortened URL.
        """
    
    def decode(self, shortUrl:str)->str:
        """Decodes a shortened URL to its original URL.
        """

import secrets
import string

class CodecRandom:
    def __init__(self):
        self.code2url = {}
        self.url2code = {}
        self.alphabet = string.ascii_letters + string.digits
        self.base = "https://tinyurl.com/"

    def _new_code(self, k=7):
        return ''.join(secrets.choice(self.alphabet) for _ in range(k))

    def encode(self, longUrl: str) -> str:
        # Return same short URL if we already encoded this longUrl
        if longUrl in self.url2code:
            return self.base + self.url2code[longUrl]
        # Generate a unique code
        code = self._new_code()
        while code in self.code2url:  # extremely rare collision
            code = self._new_code()
        self.code2url[code] = longUrl
        self.url2code[longUrl] = code
        return self.base + code

    def decode(self, shortUrl: str) -> str:
        code = shortUrl.rsplit('/', 1)[-1]
        return self.code2url.get(code, "")
    

codec = CodecRandom()
short1 = codec.encode("https://example.com/article/123")
short2 = codec.encode("https://example.com/help")

print(short1)                     # e.g., https://tinyurl.com/aZ91Lm3
print(short2)                     # e.g., https://tinyurl.com/Qp7bX0Z
print(codec.decode(short1))       # https://example.com/article/123
print(codec.decode(short2))       # https://example.com/help
```

```python
class Solution:
    def decodeMessage(self, key:str, message:str)->str:
        key_split = key.split()
        key_join = "".join(key_split)
        mapped_dict = {}
        next_code = ord('a')
        secret_message = []

        for key in key_join:
            if key not in mapped_dict:
                mapped_dict[key]=chr(next_code)
                next_code += 1
                if next_code > ord('z'):break

        for mes in message:
            if mes == ' ':
                secret_message.append(' ')
            else:
                secret_message.append(mapped_dict[mes])
        return "".join(secret_message)

s=Solution()
s.decodeMessage("eljuxhpwnyrdgtqkviszcfmabo","zwx hnfx lqantp mnoeius ycgk vcnjrdb")
```

```python
class Solution:
    def smallestNumber(self, pattern:str)->str:

        res = []
        stack = []
        i=0

        while i <= len(pattern):
            stack.append(i+1)
            if i == len(pattern) or pattern[i] == "I":
                while stack:
                    res.append(str(stack.pop()))
            i += 1
        return "".join(res)
        return res

s=Solution()
s.smallestNumber("DDDD")
```

```python
class Solution:
    def restoreString(self, s:str, indices:list[int])-> str:
        len_s = len(s)
        res = ['']*len_s
        if len_s != len(indices):
            raise ValueError("indices length must equal length of s")
        if len(set(indices))!= len_s:
            raise ValueError("indices must be a permutation (no duplicates)")
            
        for pos,ch in zip(indices,s):
            res[pos]=ch
        return "".join(res)

s=Solution()
s.restoreString("codeleet",[4,5,6,7,0,2,1,3])
```

```python
class Solution:
    def mapWordWeights(self, words:list[str],weights: list[int])->str:
        res = []
        base_a = ord('a')
        base_z = ord('z')
        for word in words:
            tot = sum(weights[(ord(w)- base_a)] for w in word)
            res.append(chr(base_z - tot % 26))
        return "".join(res)

s=Solution()
s.mapWordWeights(["abcd","def","xyz"],[5,3,12,14,1,2,3,2,10,6,6,9,7,8,7,10,8,9,6,9,9,8,3,7,7,2])
```

```python
class Solution:
    def garbageCollection(self, garbage:list[str],travel:list[int])->int:
        countGarbage = {'M': 0, 'P': 0, 'G': 0}
        count = {"M":0,"P":0,"G":0}
        pickups = 0
        
        for g in range(len(garbage)):
            for j in garbage[g]:
                count[j] += 1
                if countGarbage[j] < g:
                    countGarbage[j] = g
        for g in countGarbage:
            for l in range(countGarbage[g]):    
                pickups += travel[l]
        return pickups + sum(count.values())

s=Solution()
s.garbageCollection(["MMM","PGM","GP"],[3,10])
```

```python
class Solution:
    def sortPeople(self, names: list[str], heights: list[int]) -> list[str]:
        indexed_list = list(enumerate(heights))
        sorted_index_list = sorted(indexed_list, key=lambda x:x[1])
        result = []
        
        for i in range(len(sorted_index_list)-1, -1, -1):
            result.append(names[sorted_index_list[i][0]])
            
        return result

s = Solution()
s.sortPeople(["Mary","John","Emma"],[180,165,170])
```

```python
class Solution:
    def maxDepth(self, s:str)-> int:
        max_count = 0
        count = 0
        
        for ch in s:
            if ch == "(":
                count += 1
                if count > max_count:
                    max_count = count
            elif ch == ")":
                count -= 1
        return max_count

s=Solution()
s.maxDepth("()(())(((())()))")
```

```python
class Solution:
    def checkIfPangram(self, sentence:str) -> bool:
        alphabet = "abcdefghijklmnopqrstuvwxyz"

        for ch in alphabet:
            if ch not in sentence:
                return False
        return True

s=Solution()
s.checkIfPangram("leetcode")
```

```python
class Solution:
    def subarraySum(self, nums:list[int], k:int)->int:
        count = 0
        
        if len(nums) == 1:
            if nums[0] == k:
                count += 1
        else:
            for num in range(len(nums)-1):
                if nums[num] == k or nums[num+1] == k :
                    count += 1
                elif nums[num] + nums[num+1] == k:
                    count += 1
                    # print("three",count)
            if sum(nums) == k:
                    count += 1
        return count
s=Solution()
s.subarraySum([100,1,2,3,4], 6)
```
